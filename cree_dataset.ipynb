```python
import yfinance as yahooFinance

gold = yahooFinance.Ticker("GC=F")

df = gold.history(period="max")
```

```python
df
```

```python
df["Stock Splits"].unique()
```

```python
df["Dividends"].unique()
```

```python
df.columns
```

```python
df.reset_index(inplace=True)
```

```python
df.columns
```

```python
df.dtypes
```

```python
for i in range(len(df["Date"])):
    df["Date"][i] = pd.to_datetime(df["Date"][i]).date()
```

```python
df_gold = df.drop(['Dividends', 'Stock Splits'], axis=1)
```

```python
df_gold = df_gold.rename(columns={'Open':'Open_gold','High':'High_gold','Low':'Low_gold','Close':'Close_gold','Volume':'Volume_gold'})
```
